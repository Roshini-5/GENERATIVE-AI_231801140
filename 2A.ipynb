```python
!pip install gensim

# @title Exp_2_A (Movie Reviews Dataset)
import nltk
from nltk.corpus import movie_reviews
from gensim.models import Word2Vec

# ------------------ DOWNLOADS ------------------
nltk.download('movie_reviews')
nltk.download('punkt')

# ------------------ LOAD DATASET ------------------
# Each file is a movie review
sentences = []

for fileid in movie_reviews.fileids()[:500]:   # take 500 reviews
    words = movie_reviews.words(fileid)
    sentences.append([word.lower() for word in words])

print("Total sentences:", len(sentences))

# -------------------------------
# Function to train model
# -------------------------------
def train_model(vector_size, window_size):
    model = Word2Vec(
        sentences=sentences,
        vector_size=vector_size,
        window=window_size,
        sg=1,   # Skip-gram
        min_count=2,
        workers=4
    )
    return model

# -------------------------------
# Train models with different parameters
# -------------------------------
model1 = train_model(100, 2)
model2 = train_model(100, 5)
model3 = train_model(200, 5)

# -------------------------------
# Query words
# -------------------------------
query_words = ["movie", "good", "bad", "actor"]

# -------------------------------
# Function to print results
# -------------------------------
def show_similar_words(model, title):
    print(f"\n===== {title} =====")
    for word in query_words:
        if word in model.wv:
            print(f"\nTop 8 words similar to '{word}':")
            for similar_word, score in model.wv.most_similar(word, topn=8):
                print(f"{similar_word} ({score:.4f})")
        else:
            print(f"\n'{word}' not in vocabulary")

# -------------------------------
# Display results
# -------------------------------
show_similar_words(model1, "Vector=100, Window=2")
show_similar_words(model2, "Vector=100, Window=5")
show_similar_words(model3, "Vector=200, Window=5")
```
