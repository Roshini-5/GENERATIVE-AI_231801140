```python
import nltk
import random
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag, hmm
from nltk.probability import LidstoneProbDist

# ------------------ DOWNLOADS ------------------
nltk.download("punkt")
nltk.download("punkt_tab")
nltk.download("averaged_perceptron_tagger_eng")

# ------------------ NEW DATASET ------------------
sentences = [
    "Students prepare for campus interviews",
    "Engineers build innovative software solutions",
    "Coding skills are important for jobs",
    "Companies hire talented graduates",
    "Technology is changing the world",
    "Data analysts work with large datasets",
    "Software developers write efficient code",
    "Artificial intelligence creates smart systems"
]

# ------------------ POS TAGGING ------------------
tagged_sentences = []

for sent in sentences:
    words = word_tokenize(sent.lower())
    tagged = pos_tag(words)
    tagged_sentences.append(tagged)

print("Tagged sentences:\n")
for s in tagged_sentences:
    print(s)

# ------------------ TRAIN HMM ------------------
trainer = hmm.HiddenMarkovModelTrainer()

hmm_model = trainer.train_supervised(
    tagged_sentences,
    estimator=lambda fd, bins: LidstoneProbDist(fd, 0.1, bins)
)

# =================================================
#               SENTENCE GENERATION
# =================================================

def generate_sentence_hmm(model, max_length=15):
    sampled = model.random_sample(random.Random(), max_length)
    sentence = [word for word, tag in sampled]
    return sentence

# ------------------ GENERATE ------------------
print("\nGenerated sentences:\n")

for _ in range(5):
    print(" ".join(generate_sentence_hmm(hmm_model, 12)))
```
