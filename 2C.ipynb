```python
!pip install gensim

# @title Exp_2_C (GloVe Dataset)
import numpy as np
import gensim.downloader as api

# -------------------------------
# 1. Load pre-trained model
# -------------------------------
print("Loading model (this may take time)...")
model = api.load("glove-wiki-gigaword-100")   # CHANGED MODEL
print("Model loaded!")

# -------------------------------
# 2. Define WEAT word sets (NEW DOMAIN: TECH vs ARTS)
# -------------------------------

# Target sets
tech = ["computer", "software", "engineer", "technology", "program"]
arts = ["music", "poetry", "dance", "art", "literature"]

# Attribute sets
science = ["science", "math", "physics", "engineering", "research"]
creative = ["creative", "design", "imagination", "expression", "style"]

# -------------------------------
# 3. Cosine similarity helper
# -------------------------------
def cosine_sim(word, word_list):
    return np.mean([model.similarity(word, w) for w in word_list])

# -------------------------------
# 4. WEAT score calculation
# -------------------------------
def weat_score(X, Y, A, B):
    def s(w, A, B):
        return cosine_sim(w, A) - cosine_sim(w, B)

    return sum(s(x, A, B) for x in X) - sum(s(y, A, B) for y in Y)

# -------------------------------
# 5. Effect size calculation
# -------------------------------
def effect_size(X, Y, A, B):
    s_X = [cosine_sim(x, A) - cosine_sim(x, B) for x in X]
    s_Y = [cosine_sim(y, A) - cosine_sim(y, B) for y in Y]

    mean_diff = np.mean(s_X) - np.mean(s_Y)
    std_dev = np.std(s_X + s_Y)

    return mean_diff / std_dev

# -------------------------------
# 6. Run WEAT test
# -------------------------------
score = weat_score(tech, arts, science, creative)
effect = effect_size(tech, arts, science, creative)

# -------------------------------
# 7. Print results
# -------------------------------
print("\n--- WEAT Results ---")
print("WEAT Score:", score)
print("Effect Size:", effect)

# -------------------------------
# 8. Interpretation
# -------------------------------
if effect > 0:
    print("\nInterpretation: Bias detected (tech → science, arts → creative)")
else:
    print("\nInterpretation: No strong bias detected")
```
